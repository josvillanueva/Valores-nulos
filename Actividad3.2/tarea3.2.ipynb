```python
import pandas as pd
```

```python
#definir funciones a usar
def cuenta_valores_nulos(dataframe): 
    #por columna
    valores_nulos_cols = dataframe.isnull().sum()
    #por dataframe
    valores_nulos_df = dataframe.isnull().sum().sum()

    return("Valores nulos por columna", valores_nulos_cols,
           "valores nulos por dataframe", valores_nulos_df)
```

```python
def sustitucion_mediana(dataframe):
    #separar las numericas de las str en el dataframe 
    cuantitativas_con_nulos = dataframe.select_dtypes(include=["float64","int64","float","int"])
    #separar columnas cualitativas
    cualitativas = dataframe.select_dtypes(include=["object","datetime","category"])
    #sustitur valores nulos con 
    cuantitivas = cuantitativas_con_nulos.fillna(round(cuantitativas_con_nulos.median(), 1))
    #unir data frame
    Datos_sin_nulos = pd.concat([cuantitivas, cualitativas], axis=1)

    return(Datos_sin_nulos)
```

```python
def sustitucion_promedio(dataframe):
    #separar las numericas de las str en el dataframe 
    cuantitativas_con_nulos = dataframe.select_dtypes(include=["float64","int64","float","int"])
    #separar columnas cualitativas
    cualitativas = dataframe.select_dtypes(include=["object","datetime","category"])
    #sustitur valores nulos con 
    cuantitivas = cuantitativas_con_nulos.fillna(round(cuantitativas_con_nulos.mean(), 1))
    #unir data frame
    Datos_sin_nulos = pd.concat([cuantitivas, cualitativas], axis=1)

    return(Datos_sin_nulos)
```

```python
def sustitucion_ffill(dataframe):
    #separar las numericas de las str en el dataframe 
    cuantitativas_con_nulos = dataframe.select_dtypes(include=["float64","int64","float","int"])
    #separar columnas cualitativas
    cualitativas_con_nulos = dataframe.select_dtypes(include=["object","datetime","category"])
    #sustitur valores nulos con 
    cualitativas = cualitativas_con_nulos.fillna(method="ffill")
    #unir data frame
    Datos_sin_nulos = pd.concat([cualitativas,cuantitativas_con_nulos], axis=1)

    return(Datos_sin_nulos)
```

```python
def sustitucion_bfill(dataframe):
    #separar las numericas de las str en el dataframe 
    cuantitativas_con_nulos = dataframe.select_dtypes(include=["float64","int64","float","int"])
    #separar columnas cualitativas
    cualitativas_con_nulos = dataframe.select_dtypes(include=["object","datetime","category"])
    #sustitur valores nulos con 
    cualitativas = cualitativas_con_nulos.fillna(method="bfill")
    #unir data frame
    Datos_sin_nulos = pd.concat([cualitativas,cuantitativas_con_nulos], axis=1)

    return(Datos_sin_nulos)
```

```python
def sustitucion_string_concreto(dataframe):
    #separar las numericas de las str en el dataframe 
    cuantitativas_con_nulos = dataframe.select_dtypes(include=["float64","int64","float","int"])
    #separar columnas cualitativas
    cualitativas_con_nulos = dataframe.select_dtypes(include=["object","datetime","category"])
    #sustitur valores nulos con 
    cualitativas = cualitativas_con_nulos.fillna("calendar_date")
    #unir data frame
    Datos_sin_nulos = pd.concat([cualitativas,cuantitativas_con_nulos], axis=1)

    return(Datos_sin_nulos)
```

```python
def sustitucion_constante(dataframe):
    #separar las numericas de las str en el dataframe 
    cuantitativas_con_nulos = dataframe.select_dtypes(include=["float64","int64","float","int"])
    #separar columnas cualitativas
    cualitativas_con_nulos = dataframe.select_dtypes(include=["object","datetime","category"])
    #sustitur valores nulos con 
    cuanlitativas = cualitativas_con_nulos.fillna("f")
    #unir data frame
    Datos_sin_nulos = pd.concat([cuanlitativas,cuantitativas_con_nulos], axis=1)

    return(Datos_sin_nulos)
```

```python
def sustitucion_constante_numero(dataframe):
    #separar las numericas de las str en el dataframe 
    cuantitativas_con_nulos = dataframe.select_dtypes(include=["float64","int64","float","int"])
    #separar columnas cualitativas
    cualitativas_con_nulos = dataframe.select_dtypes(include=["object","datetime","category"])
    #sustitur valores nulos con 
    cuantitativas = cuantitativas_con_nulos.fillna(3)
    #unir data frame
    Datos_sin_nulos = pd.concat([cuantitativas,cualitativas_con_nulos], axis=1)

    return(Datos_sin_nulos)
```

```python
#Remplazar columnas tipo numeric con mediana (int y float)
#bathrooms","beds","bedrooms
sustitucion1 = sustitucion_mediana(Berlin_data[["id","bathrooms","beds","bedrooms"]])
sustitucion1
```

```python
#codigo para generar un data frame con todas las columnas y los valores nulos de las columnas ::: remplazados
df_excluido = Berlin_data.drop(columns=["bathrooms","beds","bedrooms"]) 
remplazar_con_mediana = pd.merge(sustitucion1,df_excluido, on="id")
remplazar_con_mediana
remplazar_con_mediana.to_csv("remplazar_con_mediana.csv")
```

Decide remplazar esta columnas con la media ya que necesitan ser numeros de .5 o enteros, por lo tanto remplazarlos por promedio no hubiera sido muy util. Así mismo, la mediana siempre es un numero entero, por lo que en este caso es ideal.

```python
#Remplazar columnas tipo object con ffill (object datetime y category)
#Honame y description
sustitucion2 = sustitucion_ffill(Berlin_data[["id","host_name","description"]])
sustitucion2
```

```python
#codigo para generar un data frame con todas las columnas y los valores nulos de las columnas ::: remplazados
df_excluido2 = Berlin_data.drop(columns=["host_name","description"]) 
remplazar_con_ffill = pd.merge(sustitucion2,df_excluido2, on="id")
remplazar_con_ffill.to_csv("remplazar_con_ffill.csv")
```
